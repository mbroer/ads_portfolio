```python
def recepten_bij_tag(tag):
    a = tagsdf.loc[tagsdf['tag'] == tag].recipe.to_list()
    return a
def tags_bij_recept(gerecht):
    b = tagsdf.loc[tagsdf['recipe'] == gerecht].tag.unique()
    return b
```

```python
def User_Favo_Random_Tags(randomTag = random.choices(tagsdf["tag"].to_list(), k=1), K=10):
    Favorietegerechten1 = random.choices(recepten_bij_tag(randomTag[0]), k= K)
    Favorietegerechten, Test_Favorieten = Favorietegerechten1[:K*0.8], Favorietegerechten1[K*0.8:]
    
    UserList_Tags = [tags_bij_recept(x) for x in Favorietegerechten]
    randomRecipes = random.choices(recipesdf["title"].to_list(), k=K)
    
    Random_Tags = [tags_bij_recept(x) for x in randomRecipes]

    #randomRecipes_namen = randomRecipes.copy()
    return Favorietegerechten, UserList_Tags, Random_Tags, randomRecipes, randomTag, K, Test_Favorieten

#Favorietegerechten, UserList_Tags, randomTags, randomRecipes, UserTag = User_Favo_Random_Tags()
#display(randomTags, randomRecipes)
```

```python
#Dubbele Matrix
UserList, UserList_Tags, randomTags, randomRecipes, Usertag, K, Test_Favorieten  = User_Favo_Random_Tags(K=20)
def generateDf(UserList, UserList_Tags, randomTags, randomRecipes, Usertag, K):
    list_of_tags = tagsdf['tag'].unique().tolist()
    matrix = pd.DataFrame(columns = list_of_tags)

    for i in range(K*2):
        matrix.loc[matrix.shape[0]] = 0
    #X = np.array(userList)
    X = np.array(UserList)
    loo = LeaveOneOut()
    index_counter = 0
    

    matrix2 = matrix.copy()
    matrix2.columns = ["2-" + tag for tag in matrix2.columns]
    matrix2['y'] = 0

    for train_index, test_index in loo.split(X):
        X_train, X_test = X[train_index], X[test_index]
        print("TrainList:", X_train, "One out:", X_test)

        #randomTag = getRandomValueExcludingFromList(X_train, tagList)

        randomTags0 = randomTags.pop(0)
        randomgerecht0 = randomRecipes.pop(0)
        print(randomTags0)
    
        #Tags van de gerechten binnen X_Train en X_test aangeven
#         X_train_tags = np.concatenate([tags_bij_recept(x) for x in X_train])
#         X_test_tags = np.concatenate([tags_bij_recept(x) for x in X_test])

        X_train_tags_unique = np.unique(np.concatenate([tags_bij_recept(x) for x in X_train]))
        X_test_tags_unique = np.unique(np.concatenate([tags_bij_recept(x) for x in X_test]))
        
    #----- Matrix met train vullen
        for i in range(len(X_train_tags_unique)):
            #Fill all the other data as 1's
            matrix.loc[index_counter, X_train_tags_unique[i]] = 1
            matrix.loc[index_counter+1, X_train_tags_unique[i]] = 1
            
            
        #matrix2 = matrix.copy()
        #matrix2['y'] = 0
        
    #----- Matrix2     TESTMATRIX (Rij 1 (Test) en Rij 2 (Random_recipe) los vullen in loop)
        #2e matrix invullen        
        #for i in range(len(X_test_tags_unique)):
        #    matrix2.loc[index_counter, X_test_tags_unique[i]] = 1
        #    matrix2.loc[index_counter+1, X_test_tags_unique[i]] = 1
            
        #Rij 1 
        for i in range(len(X_test_tags_unique)):
            matrix2.loc[index_counter, "2-" + X_test_tags_unique[i]] = 1
        
        #Rij 2        
        for i in range(len(randomTags0)):
            matrix2.loc[index_counter+1, "2-" + randomTags0[i]] = 1
        
        #Put the value for the random tag as 1 and put y at 0, because it should be false
        matrix2.loc[index_counter, 'y'] = 1
        matrix2.loc[index_counter+1, 'y'] = 0
        
    #----- Matrix2visueel  
        #Show which random tag is taken and the one out
        matrix2.loc[index_counter+1, 'Randomgerecht'] = str(randomgerecht0)
        matrix2.loc[index_counter, 'one out'] = X_test
        matrix2.loc[index_counter+1, 'one out'] = X_test
        
    #----- Test_Matrix

        
        #testmatrix = matrix2.drop(['y', 'randomTag', 'one out'], axis=1)
        
        #Update matrix row index counter with + 2
        index_counter += 2
    #matrix2.columns = ["2-" + tag for tag in matrix2.columns if tag =='y']
    return matrix, matrix2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.inspection import DecisionBoundaryDisplay

names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Gaussian Process",
    "Decision Tree",
    "Random Forest",
    "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "QDA",
]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
]

X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    linearly_separable,
]

figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=42
    )

    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(["#FF0000", "#0000FF"])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k")
    # Plot the testing points
    ax.scatter(
        X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, edgecolors="k"
    )
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        DecisionBoundaryDisplay.from_estimator(
            clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
        )

        # Plot the training points
        ax.scatter(
            X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k"
        )
        # Plot the testing points
        ax.scatter(
            X_test[:, 0],
            X_test[:, 1],
            c=y_test,
            cmap=cm_bright,
            edgecolors="k",
            alpha=0.6,
        )

        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(
            x_max - 0.3,
            y_min + 0.3,
            ("%.2f" % score).lstrip("0"),
            size=15,
            horizontalalignment="right",
        )
        i += 1

plt.tight_layout()
plt.show()
```
